```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data = pd.merge (study_results,mouse_metadata, on = "Mouse ID", how = "left")


# Display the data table for preview
#combined_data["Tumor Volume (mm3)"] = combined_data["Tumor Volume (mm3)"].map("{:,.1f}".format)
combined_data.head()
```

```python
# Checking the number of mice.
len(combined_data["Mouse ID"].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mouse = combined_data.loc[combined_data.duplicated(subset = ["Mouse ID", "Timepoint"]),'Mouse ID'].unique()
duplicate_mouse
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
combined_data.loc[combined_data["Mouse ID"] == "g989"]
```

```python
#identifying duplciate mouse in DF
drop_duplicate = combined_data.isin(['g989'])
drop_duplicate
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data = combined_data[drop_duplicate ["Mouse ID"] == False]
clean_data.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_data["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

mean = clean_data.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']
median = clean_data.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
variance = clean_data.groupby('Drug Regimen').var()['Tumor Volume (mm3)']
std = clean_data.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
sem = clean_data.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']

summary_stats = pd.DataFrame({
                                "Mean Tumor Volume":mean,
                                "Median Tumor Volume":median,
                                "Tumor Volume Variance":variance,
                                "Tumor Volume Std. Dev.":std,
                                "Tumor Volume Std. Err.":sem
                            })
summary_stats
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

aggregated_summary = clean_data.groupby('Drug Regimen').agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
aggregated_summary 
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_timepoint = clean_data.groupby(['Mouse ID'])['Timepoint'].max()
max_timepoint = max_timepoint.reset_index()
max_timepoint
```

```python
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
group_merge = max_timepoint.merge(clean_data,on=['Mouse ID','Timepoint'],how="left")
group_merge
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drug_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_volume_list = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in drug_list:
    
    print(f"\n----- {drug} Data ------")
   
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_vol = group_merge.loc[group_merge["Drug Regimen"] == drug, 'Tumor Volume (mm3)']
    
    
    # add subset 
    tumor_volume_list.append(tumor_vol)
    
    # Determine outliers using upper and lower bounds
    
    quartiles = tumor_vol.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    
    print(f"LowerQ is: {lowerq}")
    print(f"UpperQ is: {upperq}")
   
    print(f"   IQR is: {iqr}")
    
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f"Outlier has to be above {upper_bound} or below {lower_bound}.")
    
    outliers = tumor_vol.loc[(tumor_vol < lower_bound) | (tumor_vol > upper_bound)]
    
    if outliers.empty == False :
        print(f"{drug}'s potential outliers :{outliers}")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

plt.boxplot(tumor_volume_list, labels = drug_list , flierprops=dict(markersize=12, markerfacecolor='red'))

plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin


#find the mouse first
find_mouse = clean_data.loc[clean_data['Mouse ID']== 'l509']

#Find the required drug given to that mouse
capomulin_drug = find_mouse.loc[find_mouse['Drug Regimen'] == "Capomulin"]

plt.plot(capomulin_drug['Timepoint'],capomulin_drug['Tumor Volume (mm3)'])

plt.title('Capomulin treatment of mouse l509')
plt.xlabel('Timepoint (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

capomulin_regimen = clean_data.loc[clean_data['Drug Regimen'] == "Capomulin"]

average_everything = capomulin_regimen.groupby(['Mouse ID']).mean()

plt.scatter(average_everything['Weight (g)'],average_everything['Tumor Volume (mm3)'])

plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

x_axis = average_everything['Weight (g)']
y_axis = average_everything['Tumor Volume (mm3)']


correlation = round(st.pearsonr(x_axis, y_axis)[0],2)

print(f"The correlation between mouse weight and the average tumor volume is {correlation}")

model = st.linregress(x_axis ,y_axis)

y_values = x_axis*model[0]+model[1]

plt.scatter(x_axis,y_axis)

plt.plot(x_axis,y_values,color="red")

#plt.annotate(line_eq,(6,10),fontsize=15,color="red")

plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```
